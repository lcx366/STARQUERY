```python
def readh5_starcatalog(infile):
    '''
    infile -> [str] input file
    '''
    fin = h5py.File(infile,'r')

    # read data
    

    fp_radecs,stars_xy,stars_invariants, stars_asterisms = [],[],[]
    for j in range(48): # len(fp_radecs)
        fp_radecs.append(fin['fp_radecs/'+str(j)][:])
        stars_xy.append(fin['stars_xy/'+str(j)][:])
        stars_invariants.append(fin['stars_invariants/'+str(j)][:])
        stars_asterisms.append(fin['stars_asterisms/'+str(j)][:])
    
    # close the file
    fin.close()   
    return stars_xy,stars_invariants,stars_asterisms # fp_radecs,
```

```python
def seq2tile(seq_name,res):
    """
    Parse the name of a sequence star catalog file, and output the corresponding name of the star catalog tile file.
    
    Inputs:
    seq_name -> [str] Name of the corresponding sequence file, such as 'starcatalogs/raw/apass/apass-14.csv'   
    res -> [int] size of tile in deg

    Outputs:
    tile_name -> [str] Name of the star catalog tile file to parse, such as 'CatalogSearch.aspx?bbox=266,-90,268,-88&format=csv&catalog=apass'    
    """
    
    seq_name_sep = seq_name.split('-')
    scname = seq_name_sep[0].split('/')[-1]
    seq = int(seq_name_sep[1][:-4])  
    
    count_lon = 360//res
    ra_min = seq%count_lon*res
    ra_max = ra_min + res
    codec_min = seq//count_lon*res
    codec_max = codec_min + res
    dec_min = 90 - codec_max
    dec_max = 90 - codec_min
    tile_name = 'CatalogSearch.aspx?bbox={:d},{:d},{:d},{:d}&format=csv&catalog={:s}'.format(ra_min,dec_min,ra_max,dec_max,scname)
    
    return tile_name
```
